```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Пример данных: [средний доход, средний расход]
X = np.array([[50000, 2000], [55000, 3000], [60000, 4000],
              [250000, 50000], [300000, 60000], [350000, 70000],
              [70000, 1500], [80000, 2000], [90000, 2500]])

# Применение k-средних
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# Визуализация результатов
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.scatter(centroids[:, 0], centroids[:, 1], s=300, c='red', marker='X')
plt.title('Кластеризация методом k-средних')
plt.xlabel('Средний доход')
plt.ylabel('Средний расход')
plt.show()
```

```python
import numpy as np
from scipy.cluster import hierarchy
import matplotlib.pyplot as plt
 
# Создание массива
x = np.array([100., 200., 300., 400., 500., 250.,
              450., 280., 450., 750.])
 
# Изображение иерархической кластеризации в виде дендрограммы
temp = hierarchy.linkage(x, 'single')
plt.figure()
 
dn = hierarchy.dendrogram(
    temp, above_threshold_color="green", color_threshold=.7)
```

```python
import numpy as np
import pandas as pd
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Данные о странах: GDP per capita (тыс. долларов) и Unemployment rate (%)
data = pd.DataFrame({
    'Country': ['Country A', 'Country B', 'Country C', 'Country D', 'Country E', 'Country F'],
    'GDP_per_capita': [40, 20, 30, 80, 100, 60],
    'Unemployment_rate': [5, 7, 6, 4, 2, 3]
})

# Нормализация данных
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data[['GDP_per_capita', 'Unemployment_rate']])

# Применение агломеративной кластеризации
linked = linkage(data_scaled, method='ward')
```

```python
print(linked)
```

```python
# Построение дендрограммы
plt.figure(figsize=(10, 7))
dendrogram(linked, labels=data['Country'].values)
plt.title('Дендрограмма стран')
plt.xlabel('Страны')
plt.ylabel('Расстояние')
plt.show()
```
