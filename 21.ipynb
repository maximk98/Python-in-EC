```python
import seaborn as sns
import matplotlib.pyplot as plt

#  Пример  данных:  распределение  цен  на  недвижимость
data = [250000, 300000, 280000, 350000, 320000,
        400000, 290000, 330000, 370000, 310000]

sns.histplot(data, bins=10)
plt.title('Распределение цен на недвижимость')
plt.xlabel('Цена')
plt.ylabel('Частота')
plt.show()
```

```python
sns.kdeplot(data) # используем те же данные
plt.title('Распределение цен на недвижимость (KDE)')
plt.xlabel('Цена')
plt.ylabel('Плотность')
plt.show()
```

```python
import pandas as pd

# Доход  в  зависимости  от  уровня  образования
data = {'Образование': ['Бакалавр', 'Бакалавр', 'Магистр',
                        'Бакалавр', 'Магистр','Кандидат',
                        'Бакалавр', 'Магистр', 'Кандидат',
                        'Бакалавр'],
        'Доход': [50000, 55000, 70000, 60000, 80000,
                  100000, 52000, 75000, 120000, 58000]}
df = pd.DataFrame(data)

sns.boxplot(x='Образование', y='Доход', data=df)
plt.title('Распределение доходов по уровню образования')
plt.show()
```

```python
sns.violinplot(x='Образование', y='Доход', data=df)
plt.title('Распределение доходов по уровню образования')
plt.show()
```
