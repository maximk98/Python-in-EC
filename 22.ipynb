```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Создаем DataFrame с данными о клиентах и типе их кредитных карт
data = {'Клиент': range(1, 21),
        'Тип карты': ['Gold', 'Classic', 'Platinum', 'Classic', 'Gold', 
                     'Gold', 'Classic', 'Platinum', 'Classic', 'Gold', 
                     'Classic', 'Gold', 'Platinum', 'Classic', 'Gold',
                     'Platinum', 'Classic', 'Gold', 'Classic', 'Platinum']}
df = pd.DataFrame(data)

plt.figure(figsize=(8, 6))  # Увеличиваем размер графика для наглядности

# sns.countplot(x='Тип карты', data=df, palette='viridis') - устарело
sns.countplot(x='Тип карты', data=df, hue='Тип карты', palette='viridis')

plt.title('Распределение клиентов по типу кредитной карты')
plt.xlabel('Тип карты')
plt.ylabel('Количество клиентов')
plt.show()
```

```python
#  Пример  данных:  объем  продаж  по  месяцам
data = {'Месяц': ['Янв', 'Фев', 'Мар', 'Апр', 'Май', 'Июн',
                  'Янв', 'Фев', 'Мар', 'Апр', 'Май', 'Июн'],
        'Продажи': [15, 18, 22, 25, 30, 28,
                   17, 20, 25, 28, 32, 30],
        'Год': [2022, 2022, 2022, 2022, 2022, 2022,
               2023, 2023, 2023, 2023, 2023, 2023]}
df = pd.DataFrame(data)

sns.pointplot(x='Месяц', y='Продажи', hue='Год', data=df)
plt.title('Объем продаж по месяцам')
plt.show()
```

```python
sns.pointplot(x='Месяц', y='Продажи', hue='Год', data=df,
             markers=['o', 's'],     # маркеры
                                     # для разных лет
             linestyles=["-", "--"], # стили линий
                                     # для разных лет
             errorbar=('ci', 95))    # доверительный
                                     # интервал  95%
plt.title('Объем  продаж  по  месяцам')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Создаем DataFrame с данными о марках автомобилей и их ценах
data = {'Марка': ['Audi', 'BMW', 'Mercedes', 'Audi', 'BMW',
                  'Mercedes', 'Audi', 'BMW', 'Mercedes',
                  'Audi', 'BMW', 'Mercedes'],
        'Цена': [45000, 52000, 60000, 48000, 55000, 63000, 
                  42000, 50000, 58000, 46000, 53000, 61000]}
df = pd.DataFrame(data)
plt.figure(figsize=(8, 6))  # Увеличиваем размер графика

sns.stripplot(x='Марка', y='Цена', data=df, hue='Марка',
              jitter=0.2, size=8, palette='Set2')
plt.title('Распределение цен на автомобили разных марок')
plt.xlabel('Марка автомобиля')
plt.ylabel('Цена (USD)')
plt.show()
```

```python
# Создаем DataFrame с данными о продажах
data = {'День недели': ['Понедельник', 'Вторник', 'Среда',
                        'Четверг', 'Пятница','Суббота',
                        'Воскресенье'] * 3,
        'Категория': ['Электроника', 'Одежда', 'Продукты'] * 7,
        'Продажи': [120, 150, 130, 138, 200, 250, 230,
                   180, 200, 90, 110, 130, 160, 240,
                   200, 220, 210, 124, 260, 300, 280]}
df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))  

sns.swarmplot(x='День недели', y='Продажи', hue='Категория',
              data=df, palette='Set1')
plt.title('Распределение по дням недели и категориям')
plt.xlabel('День недели')
plt.ylabel('Количество проданных товаров')
plt.show()
```
