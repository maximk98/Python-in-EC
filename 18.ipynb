```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = {'ВВП': [40000, 50000, 35000, 60000, 45000, 70000, 80000],
        'Безработица': [5, 3, 7, 4, 6, 2.5, 2]}
df = pd.DataFrame(data)

sns.scatterplot(x='ВВП', y='Безработица', data=df)
plt.title('Взаимосвязь ВВП и уровня безработицы')
plt.xlabel('ВВП на душу населения (USD)')
plt.ylabel('Уровень безработицы (%)')
plt.show()
```

```python
data = {'ВВП': [40000, 50000, 35000, 60000, 45000, 70000, 80000],
        'Безработица': [5, 3, 7, 4, 6, 2.5, 2],
        'Регион': ['Азия', 'Европа', 'Азия', 'Европа', 'Азия',
                   'Европа', 'Америка']}
df = pd.DataFrame(data)

sns.scatterplot(x='ВВП', y='Безработица', hue='Регион', data=df)
plt.title('Взаимосвязь ВВП и уровня безработицы по регионам')
plt.xlabel('ВВП на душу населения (USD)')
plt.ylabel('Уровень безработицы (%)')
plt.show()
```

```python
sns.scatterplot(x='ВВП', y='Безработица', 
                hue='Регион', 
                style='Регион',  #  разные  маркеры  для  регионов
                size='ВВП',     #  размер  точек  зависит  от  ВВП
                sizes=(30, 300), #  диапазон  размеров  точек
                palette='viridis', #  цветовая  палитра
                data=df)
plt.title('Взаимосвязь  ВВП  и  уровня  безработицы  по  регионам')
plt.xlabel('ВВП  на  душу  населения  (USD)')
plt.ylabel('Уровень  безработицы  (%)')
plt.show()
```

```python
data = {'Год': [2018, 2019, 2020, 2021, 2022],
        'ИПЦ': [100, 105, 110, 118, 125]}
df = pd.DataFrame(data)

sns.lineplot(x='Год', y='ИПЦ', data=df, marker='o')
plt.title('Динамика индекса потребительских цен')
plt.xlabel('Год')
plt.ylabel('ИПЦ (базисный год = 100)')
plt.show()
```

```python
data = {'Год': [2018, 2019, 2020, 2021, 2022, 2018, 2019, 2020, 2021, 2022],
        'Значение': [100, 105, 110, 118, 125, 90, 95, 100, 108, 115],
        'Показатель': ['ИПЦ', 'ИПЦ', 'ИПЦ', 'ИПЦ', 'ИПЦ', 'ПМ', 'ПМ', 'ПМ',
                       'ПМ', 'ПМ']}
df = pd.DataFrame(data)

sns.lineplot(x='Год', y='Значение', hue='Показатель', data=df, marker='o')
plt.title('Динамика  ИПЦ  и  прожиточного  минимума')
plt.xlabel('Год')
plt.ylabel('Значение (базисный год = 100)')
plt.show()
```

```python
sns.lineplot(x='Год', y='Значение', hue='Показатель',
             style='Показатель',  #  разные  стили  линий
             markers=True,        #  отображаем  маркеры  для  всех  линий
             dashes=False,         #  делаем  все  линии  сплошными
             palette='Set1',       #  цветовая  палитра
             data=df)
plt.title('Динамика  ИПЦ  и  прожиточного  минимума')
plt.xlabel('Год')
plt.ylabel('Значение  (базисный  год  =  100)')
plt.grid(True) #  добавляем  сетку
plt.show()
```
