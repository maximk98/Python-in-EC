```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

#  Пример данных:  доход  по  отраслям
data = {'Отрасль': ['IT', 'Финансы', 'Промышленность', 'Образование'],
        'Доход': [70, 65, 55, 45]}
df = pd.DataFrame(data)

#  Столбчатая диаграмма с палитрой "magma"
sns.barplot(hue='Отрасль', y='Доход', data=df, palette='magma')
plt.title('Средний доход по отраслям')
plt.show()
```

```python
# Примерные данные
categories = ['Category A', 'Category B', 'Category C', 'Category D',
              'Category E', 'Category F', 'Category G', 'Category H'
             , 'Category I']
values = [5, 7, 8, 6, 10, 9, 2, 5.8, 4.5]

# Создание градиентной палитры
palette = sns.color_palette("coolwarm", 10)
custom_palette = sns.color_palette(palette, n_colors=len(values))

# Создание графика
sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))
sns.barplot(hue=categories, y=values, palette=custom_palette)

plt.title('Столбчатая диаграмма с градиентной палитрой')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
custom_palette = ['#4c72b0', '#55a868', '#c44e52', '#8172b3'] 
#  Создаем список из шестнадцатеричных цветовых кодов

sns.barplot(hue='Отрасль', y='Доход', data=df, palette=custom_palette)
plt.title('Средний  доход  по  отраслям  (ваша  палитра)')
plt.show()
```

```python
sns.set_style("darkgrid")  #  Применяем  стиль  "darkgrid"

sns.barplot(hue='Отрасль', y='Доход', data=df)
plt.title('Средний доход по отраслям')
plt.show()
```

```python
sns.set_theme(style="darkgrid", palette="muted", font='Arial', font_scale=1.2)
# Устанавливаем  тему  с  темным  фоном,  приглушенной  палитрой,
# шрифтом  Arial  и  увеличенным  размером  шрифта

sns.barplot(hue='Отрасль', y='Доход', data=df)
plt.title('Средний доход по отраслям')
plt.show()
```

```python
#  Контекст  для  презентации  с  крупными  шрифтами
sns.set_context("talk")  
sns.barplot(hue='Отрасль', y='Доход', data=df)
plt.title('Средний доход по отраслям')
plt.show()
```

```python
sns.set_theme(style="whitegrid", palette="tab10", font_scale=1.1)

#  Пример  данных:  ВВП,  инфляция  и  безработица
data = {'Год': [2018, 2019, 2020, 2021],
        'ВВП': [2.3, 2.1, -0.5, 2.7],
        'Инфляция': [2.9, 2.5, 3.2, 4.1],
        'Безработица': [5.2, 4.8, 6.5, 5.9]}
df = pd.DataFrame(data)

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

sns.lineplot(x='Год', y='ВВП', data=df, ax=axes[0], marker='o')
sns.lineplot(x='Год', y='Инфляция', data=df, ax=axes[1], marker='o')
sns.barplot(x='Год', y='Безработица', data=df, ax=axes[2])

plt.suptitle('Основные  экономические  показатели', fontsize=16)
plt.show()
```
