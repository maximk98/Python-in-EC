```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Пример данных: корреляция между экономическими показателями
data = {'ВВП': [2.3, 2.1, -0.5, 2.7],
        'Инфляция': [2.9, 2.5, 3.2, 4.1],
        'Безработица': [5.2, 4.8, 6.5, 5.9],
        'Инвестиции': [1.5, 1.8, 1.2, 1.6]}
df = pd.DataFrame(data)

corr_matrix = df.corr()  #  Вычисляем корреляционную матрицу
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Корреляционная матрица экономических показателей')
plt.show()
```

```python
import numpy as np
# Шаг 1: Генерация данных
np.random.seed(42)
years = [2018, 2019, 2020, 2021]
months = ['Янв', 'Фев', 'Мар', 'Апр', 'Май', 'Июн', 'Июл', 'Авг', 'Сен', 'Окт', 'Ноя', 'Дек']
regions = ['Регион A', 'Регион B', 'Регион C']

# Создаем DataFrame с данными о продажах
data = []
for year in years:
    for month in months:
        for region in regions:
            sales = np.random.randint(50, 150)  # случайные продажи от 50 до 150
            data.append([year, month, region, sales])
df = pd.DataFrame(data, columns=['Год', 'Месяц', 'Регион', 'Продажи'])
# Шаг 2: Преобразование данных в сводную таблицу
pivot_table = df.pivot_table(index='Месяц', columns='Год', values='Продажи', aggfunc='sum')
# Шаг 3: Визуализация данных с помощью heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(pivot_table, annot=True, fmt='g', cmap='Blues', linewidths=.5)
plt.title('Тепловая карта продаж за месяц по годам')
plt.xlabel('Год')
plt.ylabel('Месяц')
plt.show()
```

```python
# Пример данных: доходность инвестиций по регионам и типам активов
data = {
    'Регион': ['Европа', 'Европа', 'Азия', 'Азия', 'Америка', 'Америка'],
    'Актив': ['Акции', 'Облигации', 'Акции', 'Облигации', 'Акции', 'Облигации'],
    'Доходность': [8, 5, 10, 4, 6, 3]
}

df = pd.DataFrame(data)

# Создание сводной таблицы
pivot_table = df.pivot_table(index='Регион', columns='Актив', values='Доходность', aggfunc='mean', fill_value=0)

# Визуализация тепловой карты
plt.figure(figsize=(8, 6))
sns.heatmap(pivot_table, annot=True, cmap='YlGnBu', cbar_kws={'label': 'Доходность (%)'})
plt.title('Доходность инвестиций по регионам и типам активов')
plt.xlabel('Актив')
plt.ylabel('Регион')
plt.show()
```

```python
sns.heatmap(pivot_table, annot=True, fmt=".1f", cmap='coolwarm', 
            linewidths=1, linecolor='white', cbar_kws={'label': 'Доходность (%)'})
plt.title('Доходность инвестиций по регионам и типам активов')
plt.show()
```
