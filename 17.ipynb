```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Создаем список месяцев
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# Генерируем случайные данные о продажах с трендом и сезонностью
np.random.seed(42) # для воспроизводимости результатов
sales = [100 + i*5 + 50*np.sin(i*np.pi/6) + np.random.normal(0, 10) for i in range(12)]

# Создаем DataFrame
df = pd.DataFrame({'month': months, 'sales': sales})

plt.figure(figsize=(10, 6))
sns.lineplot(x='month', y='sales', data=df, marker='o', linestyle='--', color='skyblue')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# --- Генерируем данные ---
np.random.seed(42)  # Для воспроизводимости
data = np.random.normal(loc=50, scale=10, size=1000)  # 1000 значений с нормальным распределением

# --- Строим гистограмму ---
sns.histplot(data, bins=20, kde=True, color='skyblue')
plt.title('Гистограмма случайных данных')
plt.xlabel('Значение')
plt.ylabel('Частота')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# --- Генерируем данные ---
np.random.seed(42) 
n = 100
x = np.random.rand(n) * 10  
y = x + np.random.randn(n) * 2 

# --- Строим диаграмму рассеяния ---
sns.scatterplot(x=x, y=y)
plt.title('Диаграмма рассеивания')
plt.xlabel('Значения X')
plt.ylabel('Значения Y')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# --- Генерируем данные ---
np.random.seed(42)
data = np.random.rand(10, 10)

# --- Строим тепловую карту ---
plt.figure(figsize=(8, 6))
sns.heatmap(data, annot=True, cmap="coolwarm", cbar_kws={'label': 'Значение'})
plt.title('Тепловая карта')
plt.xlabel('Ось X')
plt.ylabel('Ось Y')
plt.show()
```
