```python
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка данных
data = pd.read_csv(r'd:\Mall_Customers.csv')
print(data.head())
```

```python
X = data[['Annual Income (k$)', 'Spending Score (1-100)']]

# Применение KMeans для сегментации клиентов
kmeans = KMeans(n_clusters=5, random_state=42)
data['Cluster'] = kmeans.fit_predict(X)
```

```python
# Визуализация результатов
plt.scatter(X['Annual Income (k$)'], X['Spending Score (1-100)'], c=data['Cluster'], cmap='viridis')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('Customer Segments')
plt.show()
```

```python
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# Загрузка данных
data = pd.read_csv(r'd:\Wholesale customers data.csv')
print(data.head())

# Отбор признаков для кластеризации
X = data[['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicassen']]

# Нормализация данных
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Применение KMeans для кластеризации
kmeans = KMeans(n_clusters=4, random_state=42)
data['Cluster'] = kmeans.fit_predict(X_scaled)
print(data.head())
```

```python
# Визуализация кластеров
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=data['Cluster'], cmap='viridis')
plt.xlabel('Fresh (Scaled)')
plt.ylabel('Milk (Scaled)')
plt.title('Кластеры клиентов')
plt.colorbar()
plt.show()
```
