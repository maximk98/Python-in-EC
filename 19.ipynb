```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

#  Пример  данных:  распределение  доходов  населения
np.random.seed(42)
data = np.random.normal(loc=50000, scale=15000, size=1000)

sns.histplot(data, bins=20)
plt.title('Распределение доходов населения')
plt.xlabel('Доход')
plt.ylabel('Частота')
plt.show()
```

```python
import pandas as pd
#  Пример  данных:  зарплаты  мужчин  и  женщин  в  одной  профессии
data = {'Зарплата': [45000, 55000, 50000, 60000, 58000, 48000, 52000, 57000, 62000, 65000],
        'Пол': ['Мужской', 'Мужской', 'Женский', 'Мужской', 'Женский', 'Женский', 'Мужской', 'Женский', 'Мужской', 'Женский']}
df = pd.DataFrame(data)

sns.histplot(data=df, x='Зарплата', hue='Пол', kde=True) # kde - отображение кривой плотности
plt.title('Распределение зарплат')
plt.xlabel('Зарплата')
plt.ylabel('Частота')
plt.show()
```

```python
import pandas as pd
#  Пример  данных:  зарплаты  мужчин  и  женщин  в  одной  профессии
data = {'Зарплата': [45000, 55000, 50000, 60000, 58000, 48000, 52000,
                     57000, 62000, 65000],
        'Пол': ['Мужской', 'Мужской', 'Женский', 'Мужской', 'Женский',
                'Женский', 'Мужской', 'Женский', 'Мужской', 'Женский']}
df = pd.DataFrame(data)

sns.histplot(data=df, x='Зарплата', hue='Пол', kde=True)
plt.title('Распределение зарплат')
plt.xlabel('Зарплата')
plt.ylabel('Частота')
plt.show()
```

```python
# изменение цветовой палитры и прозрачности
# изменение цветовой палитры и прозрачности
# изменение толщины линий
sns.histplot(data=df, x='Зарплата', hue='Пол', kde=True, element='step',
             palette='Set2', alpha=0.7, linewidth=2)

plt.title('Распределение  зарплат  по  полу')
plt.xlabel('Зарплата')
plt.ylabel('Частота')
plt.show()
```

```python
#  Пример данных: средний доход по отраслям
data = {'Отрасль': ['IT', 'Финансы', 'Промышленность', 'Образование'],
        'Доход': [70000, 65000, 55000, 45000]}
df = pd.DataFrame(data)

sns.barplot(x='Отрасль', y='Доход', data=df)
plt.title('Средний доход по отраслям')
plt.xlabel('Отрасль')
plt.ylabel('Средний доход')
plt.show()
```

```python
#  Пример  данных:  средний  доход  по  отраслям  и  полу
data = {'Отрасль': ['IT', 'IT', 'Финансы', 'Финансы', 'Промышленность',
                    'Промышленность', 'Образование', 'Образование'],
        'Доход': [75000, 60000, 70000, 60000, 60000, 50000, 50000, 40000],
        'Пол': ['Мужской', 'Женский', 'Мужской', 'Женский', 'Мужской',
                'Женский', 'Мужской', 'Женский']}
df = pd.DataFrame(data)

sns.barplot(x='Отрасль', y='Доход', hue='Пол', data=df)
plt.title('Средний  доход  по  отраслям  и  полу')
plt.xlabel('Отрасль')
plt.ylabel('Средний  доход')
plt.show()
```

```python
data = {'Отрасль': ['IT', 'IT', 'Финансы', 'Финансы', 'Промышленность',
                    'Промышленность', 'Образование', 'Образование'],
        'Доход': [75000, 60000, 70000, 60000, 60000, 50000, 50000, 40000],
        'Пол': ['Мужской', 'Женский', 'Мужской', 'Женский', 'Мужской',
                'Женский', 'Мужской', 'Женский']}
df = pd.DataFrame(data)

sns.barplot(x='Отрасль', y='Доход', hue='Пол', data=df, palette='pastel',
            dodge=False)
           
plt.title('Средний  доход  по  отраслям  и  полу')
plt.xlabel('Отрасль')
plt.ylabel('Средний  доход')
plt.show()
```
